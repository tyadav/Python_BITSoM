```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## 1) Matplotlib mental model: figure, axes, artists

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y, label="sin(x)")
ax.set_title("Line plot on Axes")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()
fig.tight_layout()
plt.show()
```

## 2) Seaborn mental model: data + mappings

```python
df = pd.DataFrame({
    "time": x,
    "value": y,
    "group": np.where(x < 5, "A", "B")
})
df.head()
```

```python
sns.lineplot(data=df, x="time", y="value", hue="group")
plt.title("Seaborn lineplot with hue")
plt.show()
```

```python
sns.scatterplot(data=df, x="time", y="value", hue="group")
plt.title("Seaborn scatterplot")
plt.show()
```

## 3) Distribution plots

```python
vals = np.random.normal(loc=0, scale=1, size=200)
plt.hist(vals, bins=20, color="steelblue", edgecolor="black")
plt.title("Matplotlib histogram")
plt.show()
```

```python
sns.histplot(vals, bins=20, kde=True)
plt.title("Seaborn histplot with KDE")
plt.show()
```

```python
cat_df = pd.DataFrame({
    "dept": ["Eng", "Eng", "Sales", "Sales", "HR", "HR"],
    "salary": [90, 110, 70, 75, 60, 65]
})
sns.boxplot(data=cat_df, x="dept", y="salary")
plt.title("Seaborn boxplot")
plt.show()
```

## 4) Categorical comparisons

```python
categories = ["A", "B", "C"]
values = [5, 3, 6]
plt.bar(categories, values, color="orange")
plt.title("Matplotlib bar chart")
plt.show()
```

```python
sns.barplot(x=categories, y=values)
plt.title("Seaborn barplot")
plt.show()
```

## 5) Matrix plot (heatmap)

```python
corr_df = pd.DataFrame({
    "a": np.random.rand(30),
    "b": np.random.rand(30),
    "c": np.random.rand(30)
})
corr = corr_df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation heatmap")
plt.show()
```

## 6) Exploration (pairplot)

```python
pair_df = pd.DataFrame({
    "x": np.random.normal(0, 1, 60),
    "y": np.random.normal(0, 1, 60),
    "z": np.random.normal(0, 1, 60),
    "label": np.random.choice(["A", "B"], size=60)
})
sns.pairplot(pair_df, hue="label")
plt.show()
```
